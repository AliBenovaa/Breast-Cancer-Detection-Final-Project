```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the bar chart
models = ('CNN', 'VGG16', 'InceptionV3', 'ResNet')
accuracy = [0.9, 0.9, 0.84, 0.88]
time = [82, 34, 58, 47]

# Define the colors for the bars
colors = ['green', 'orange', 'blue', 'red']

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Plot the accuracy data on the first subplot
ax1.bar(models, accuracy, color=colors)
ax1.set_title('Accuracy')
ax1.set_ylim([0.8, 1.0])
ax1.set_ylabel('Accuracy')

# Plot the time data on the second subplot
ax2.bar(models, time, color=colors)
ax2.set_title('Time')
ax2.set_ylim([0, 100])
ax2.set_ylabel('Time (seconds)')

# Adjust the layout and save the figure
fig.tight_layout()
plt.savefig('performance.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
models = ('CNN', 'VGG16', 'InceptionV3', 'ResNet')
accuracy = [0.9, 0.9, 0.84, 0.88]
time = [82, 34, 58, 47]

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Plot the accuracy data on the first subplot
ax1.boxplot([accuracy])
ax1.set_xticklabels(['Accuracy'])
ax1.set_ylim([0.8, 1.0])
ax1.set_ylabel('Accuracy')

# Plot the time data on the second subplot
ax2.boxplot([time])
ax2.set_xticklabels(['Time'])
ax2.set_ylim([0, 100])
ax2.set_ylabel('Time (seconds)')

# Add the x-tick labels
plt.setp((ax1, ax2), xticks=[1])

# Set the colors
box_colors = ['green', 'orange', 'blue', 'red']
for i, box in enumerate(ax1.artists):
    box.set_facecolor(box_colors[i])
for i, box in enumerate(ax2.artists):
    box.set_facecolor(box_colors[i])

# Adjust the layout and save the figure
fig.tight_layout()
plt.savefig('performance_boxplot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

epochs = range(1, 11)
times = [76, 64, 58, 58, 59, 64, 59, 58, 65, 59]
accuracies = [0.8431, 0.8468, 0.8468, 0.8469, 0.8466, 0.8467, 0.8466, 0.8465, 0.8463, 0.8464]

fig, ax1 = plt.subplots(figsize=(10, 6))

# plot the time taken for each epoch
ax1.bar(epochs, times, color='#1f77b4') # blue color
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Time (s)', color='#1f77b4')
ax1.tick_params(axis='y', labelcolor='#1f77b4')
ax1.set_ylim(0, 100)

# plot the accuracy for each epoch
ax2 = ax1.twinx()
ax2.plot(epochs, accuracies, marker='o', color='#ff7f0e') # orange color
ax2.set_ylabel('Accuracy', color='#ff7f0e')
ax2.tick_params(axis='y', labelcolor='#ff7f0e')
ax2.set_ylim(0.84, 0.85)

plt.title('Model Speed and Accuracy')
plt.show()
```
